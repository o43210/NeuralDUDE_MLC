```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_dude_result = np.zeros(16)
n_dude_delta_result = np.zeros(16)

n_dude_em_result = np.zeros(16)
n_dude_em_delta_result = np.zeros(16)

cude_result = np.zeros(16)

BW_result = np.zeros(16)
FB_result = np.zeros(16)


delta = 0.1

BW_result += 0.05652
FB_result += 0.056519


with open("NeuralDUDE_k_result.txt", "r") as f:
    for i in xrange(16):
        value = float(f.readline())
        n_dude_result[i] = value

with open("NeuralDUDE_delta_result.txt", "r") as f:
    for i in xrange(16):
        value = float(f.readline())
        n_dude_delta_result[i] = value
opt_delta = float(np.argmin(n_dude_delta_result)) / float(100)

with open("CUDE_result.txt", "r") as f:
    for i in xrange(16):
        value = float(f.readline())
        cude_result[i] = value

with open("NeuralDUDE_EM_v2_result_error.txt", "r") as f:
    for i in xrange(16):
        value = float(f.readline())
        n_dude_em_result[i] = value
opt_em_k = np.argmin(n_dude_em_result)

with open("NeuralDUDE_EM_v2_result_delta.txt", "r") as f:
    for i in xrange(16):
        value = float(f.readline())
        n_dude_em_delta_result[i] = value
    
```

```python
n_dude_result = n_dude_result / delta
n_dude_delta_result = n_dude_delta_result / delta
cude_result = cude_result / delta
n_dude_em_result = n_dude_em_result / delta
BW_result =BW_result / delta
FB_result = FB_result / delta


print('n_dude_result:')
print(n_dude_result)
print('n_dude_delta_result:')
print(n_dude_delta_result)
print('cude_result:')
print(cude_result)
print('n_dude_em_result')
print(n_dude_em_result)
print('n_dude_em_delta_result')
print(n_dude_em_delta_result)
print('BW_result:')
print(BW_result)
print('FB_result:')
print(FB_result)
```
